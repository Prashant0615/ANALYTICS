### Step 13: Handled Outliers in Numerical Columns Using the IQR Method

- In this step, outliers in key numerical columns were addressed to improve data quality and ensure robust analysis.
- The Interquartile Range (IQR) method was used to detect and remove outliers. The IQR is calculated as the difference between the 75th percentile (Q3) and the 25th percentile (Q1) of a dataset.
- For each numerical column (`sellingprice`, `odometer`, `condition`, `car_age`), the following process was applied:
    - Compute Q1 (25th percentile) and Q3 (75th percentile).
    - Calculate the IQR as Q3 - Q1.
    - Determine the lower bound as Q1 - 1.5 * IQR and the upper bound as Q3 + 1.5 * IQR.
    - Filter out rows where the column value falls outside these bounds, as these are considered outliers.
- This method helps to remove extreme values that could skew statistical analyses and visualizations, leading to more accurate and reliable insights.
- After outlier removal, the DataFrame was reset to ensure a continuous index.
- Handling outliers is crucial for many machine learning algorithms and statistical models, as it prevents undue influence from anomalous data points.

```python
print(df.describe()) # for numerical columns
```

```python
# Handle outliers using the IQR method for numerical columns
numeric_cols = ['sellingprice', 'odometer', 'condition', 'car_age']

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Filter out outliers
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

df.reset_index(drop=True, inplace=True)
df.describe()
```
